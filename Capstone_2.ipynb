**Overview Company :** The New York City Taxi and Limousine Commission (TLC) 

The New York City Taxi and Limousine Commission (TLC), created in 1971, is the agency responsible for licensing and regulating New York City's Medallion (Yellow) taxi cabs, for-hire vehicles (community-based liveries, black cars and luxury limousines), commuter vans, and paratransit vehicles. The Commission's Board consists of nine members, eight of whom are unsalaried Commissioners. The salaried Chair/ Commissioner presides over regularly scheduled public commission meetings and is the head of the agency, which maintains a staff of approximately 600 TLC employees.

Over 200,000 TLC licensees complete approximately 1,000,000 trips each day. To operate for hire, drivers must first undergo a background check, have a safe driving record, and complete 24 hours of driver training. TLC-licensed vehicles are inspected for safety and emissions at TLC's Woodside Inspection Facility.

TLC licenses over 130,000 vehicles in New York City. Each vehicle receives comprehensive safety and emissions inspections by TLC and must be driven by TLC-licensed drivers that have undergone background checks and passed TLC education requirements. ## Outliers

```python
def count_outliers_with_percentage(df, numeric_cols):
    """
    Counts the total number of outliers and their percentage in specified numeric columns of a DataFrame.

    Args:
        df: The input pandas DataFrame.
        numeric_cols: A list of column names to check for outliers.

    Returns:
        A dictionary where keys are column names and values are tuples containing 
        the number of outliers and their percentage in each column.
    """

    outlier_counts = {}
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        count_outliers = len(outliers)
        percentage_outliers = (count_outliers / len(df)) * 100
        outlier_counts[col] = (count_outliers, round(percentage_outliers, 2))
    
    total_outliers = sum([val[0] for val in outlier_counts.values()])
    print("Total number of outliers across all specified columns:", total_outliers)
    return outlier_counts

# Example usage (assuming 'numerik' DataFrame from previous code):
numeric_cols = numerik.columns  # Replace with your actual numeric columns
outlier_counts = count_outliers_with_percentage(df, numeric_cols)

print("\nNumber of outliers and percentage per column:")
for col, (count, percentage) in outlier_counts.items():
    print(f"{col}: {count} outliers ({percentage}%)")
```

## Time Grouping

```python
# Create a new column for the month name
df['month'] = df['lpep_pickup_datetime'].dt.month

# Create a new column for the day of the month
df['day'] = df['lpep_pickup_datetime'].dt.day

# Create a new column for hour of the day
df['hour'] = df['lpep_pickup_datetime'].dt.hour
```

```python
# Mendefinisikan fungsi untuk mengelompokkan waktu
def time_group(hour):
    if 0 <= hour < 6:
        return 'Dawn'  # Jam 00:00-05:59 (Pagi fajar atau dini hari)
    elif 6 <= hour < 12:
        return 'Morning'  # Jam 06:00-11:59
    elif 12 <= hour < 18:
        return 'Afternoon'  # Jam 12:00-17:59
    elif 18 <= hour < 21:
        return 'Evening'  # Jam 18:00-20:59
    else:
        return 'Night'  # Jam 21:00-23:59

# Membuat kolom baru 'time_group' berdasarkan jam pada 'lpep_pickup_datetime'
df['time_group'] = df['lpep_pickup_datetime'].dt.hour.apply(time_group)
```

### Proportional Analysis

```python
import matplotlib.pyplot as plt

# Fungsi untuk menggabungkan kategori kecil menjadi "Other"
def combine_small_categories(data, column, threshold=5):
    # Hitung persentase setiap kategori
    counts = data[column].value_counts()
    total_count = counts.sum()
    percentages = counts / total_count * 100

    # Pisahkan kategori kecil
    small_categories = percentages[percentages < threshold]
    large_categories = percentages[percentages >= threshold]

    # Gabungkan kategori kecil menjadi "Other"
    combined = large_categories.copy()
    if not small_categories.empty:
        combined['Other'] = small_categories.sum()

    return combined

# Fungsi untuk membuat pie chart dengan lingkaran di tengah (doughnut chart)
def create_pie_chart(data, title, ax):
    labels = data.index
    sizes = data.values
    total_count = df_2023[column].value_counts().sum()

    # Membuat pie chart
    wedges, texts, autotexts = ax.pie(
        sizes, 
        labels=labels, 
        autopct='%1.2f%%', 
        startangle=140, 
        explode=(0.1,) + (0,) * (len(labels) - 1),  # Eksplosi kategori pertama
        colors=plt.cm.Paired.colors, 
        pctdistance=0.8
    )

    # Menambahkan lingkaran di tengah
    centre_circle = plt.Circle((0, 0), 0.6, fc='white')
    ax.add_artist(centre_circle)

    # Menambahkan judul
    ax.set_title(title, fontsize=14, fontweight='bold')

    # Menambahkan total di tengah lingkaran
    ax.text(0, 0, f"Total: {int(total_count)}", ha='center', va='center', fontsize=10)

# Gabungkan kategori kecil hanya untuk 'RatecodeID' dan 'payment_type'
ratecode_data = combine_small_categories(df_2023, 'RatecodeID')
payment_data = combine_small_categories(df_2023, 'payment_type')
triptype_data = df_2023['trip_type'].value_counts()  # Gunakan data asli tanpa perubahan

# Buat subplots untuk masing-masing kolom
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Subplot untuk 'RatecodeID'
create_pie_chart(ratecode_data, 'RatecodeID Proportion', axes[0])

# Subplot untuk 'payment_type'
create_pie_chart(payment_data, 'Payment Type Proportion', axes[1])

# Subplot untuk 'trip_type'
create_pie_chart(triptype_data, 'Trip Type Proportion', axes[2])

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Fungsi untuk membuat pie chart dengan lingkaran di tengah (doughnut chart)
def create_pie_chart(data, title, ax):
    labels = data.index
    sizes = data.values
    total_count = sum(sizes)

    # Menyesuaikan explode agar panjangnya sama dengan jumlah kategori
    explode = (0.1,) * len(labels)  # Jika ada lebih dari 1 kategori, explode untuk semua kategori

    # Membuat pie chart
    wedges, texts, autotexts = ax.pie(
        sizes, 
        labels=labels, 
        autopct='%1.2f%%', 
        startangle=140, 
        explode=explode,  # Sesuaikan explode dengan jumlah kategori
        colors=plt.cm.Paired.colors, 
        pctdistance=0.8
    )

    # Menambahkan lingkaran di tengah
    centre_circle = plt.Circle((0, 0), 0.6, fc='white')
    ax.add_artist(centre_circle)

    # Menambahkan judul
    ax.set_title(title, fontsize=14, fontweight='bold')

    # Menambahkan total di tengah lingkaran
    ax.text(0, 0, f"Total: {int(total_count)}", ha='center', va='center', fontsize=10)

# Filter hanya untuk trip type 'Dispatch' dan 'Street Hail' dengan payment_type 'Credit Card' dan 'Cash'
filtered_dispatch = df_2023[(df_2023['trip_type'] == 'Dispatch') & 
                            (df_2023['payment_type'].isin(['Credit Card', 'Cash']))]

filtered_street_hail = df_2023[(df_2023['trip_type'] == 'Street_hail') & 
                                (df_2023['payment_type'].isin(['Credit Card', 'Cash']))]

# Hitung persentase pembayaran untuk masing-masing trip type
dispatch_payment_data = filtered_dispatch['payment_type'].value_counts()
street_hail_payment_data = filtered_street_hail['payment_type'].value_counts()

# Buat subplots untuk masing-masing kategori
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Subplot untuk Dispatch
create_pie_chart(dispatch_payment_data, 'Payment Type Distribution for Dispatch', axes[0])

# Subplot untuk Street Hail
create_pie_chart(street_hail_payment_data, 'Payment Type Distribution for Street Hail', axes[1])

# Menyesuaikan layout
plt.tight_layout()
plt.show()

```

The proportion of 'Credit Card' and 'Cash' payments for both dispatch and street hail trips indicates that 'Credit Card' is the more dominant payment method in both categories.
